# D-Wave

QLauncher is compatible with D-wave backends and many d-wave based algorithm.

Let's start by defining a problem to be solved, we will use Exact Cover in this tutorial.

```python
from qlauncher.problems import EC

pr = EC.from_preset("default")
pr.visualize()
```

## D-wave Solver
D-wave solver is an algorithm dedicated to work on d-wave computers, let's initialize it:

```python
from qlauncher.routines.dwave_routines import DwaveSolver

alg = DwaveSolver()
```

QLauncher features local backends such as SimulatedAnnealingBackend, based on the simulated annealing heuristic algorithm, and TabuBackend based, on the Tabu search algorithm. There is also the DwaveBackend which allows to connect to the real D-wave computer.

```python
from qlauncher.routines.dwave_routines import DwaveBackend, SimulatedAnnealingBackend, TabuBackend

backend = SimulatedAnnealingBackend()
```

Now we can use QLauncher to launch D-wave solver on the selected backend 

```python
from qlauncher.launcher import QLauncher

launcher = QLauncher(pr,alg,backend)
result = launcher.run()
result
```

The results correspond to the following matching, since the solver is non-deterministic the result might be suboptimal:

```python
pr.visualize(result.best_bitstring)
```
