# Orca
Quantum Launcher is compatible with Orca Computing photonic computers and features algorithms based based on photonic computing

Let's start by defining a problem o be solved, we will use Exact Cover in this tutorial

```python
from quantum_launcher.problems import EC

pr = EC.from_preset("toy")
pr.visualize()
```

## Binary Bosonic Solver

Binary Bosonic solver is an algorithm dedicated to work on photonic computers just like the one from Orca computing, Let's initialize it:

```python
from quantum_launcher.routines.orca_routines import BBS

alg = BBS()
```

Now we need to define the backend our algorithm will run on, we will use a local Orca simulator, but you can also connect to the real Orca computer if you have access to one

```python
from quantum_launcher.routines.orca_routines import OrcaBackend

backend = OrcaBackend("local_simulator")
```

Now we can use Quantum launcher to launch BBS on the selected backend 

```python
from quantum_launcher.launcher import QuantumLauncher

launcher = QuantumLauncher(pr,alg,backend)
result = launcher.run()
result
```

The results corresponds to the following matching, since BBS is non-deterministic the result might be suboptimal:

```python
pr.visualize(result.best_bitstring)
```
