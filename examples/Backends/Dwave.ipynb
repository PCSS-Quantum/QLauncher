# D-Wave

Quantum Launcher is compatible with D-wave backends and many d-wave based algorithms 

Let's start by defining a problem to be solved, we will use Exact Cover in this tutorial

```python
from quantum_launcher.problems import EC

pr = EC.from_preset("toy",onehot="quadratic")
pr.visualize()
```

## D-wave Solver
D-wave solver is an algorithm dedicated to work on d-wave computers, let's initialize it:

```python
from quantum_launcher.routines.dwave_routines import DwaveSolver

alg = DwaveSolver()
```

Quantum Launcher features local backends such as SimulatedAnnealingBackend, based on the simulated annealing heuristic algorithm, and TabuBackend based, on the Tabu search algorithm. There is also the DwaveBackend which allows to connect to the real D-wave computer

```python
from quantum_launcher.routines.dwave_routines import DwaveBackend, SimulatedAnnealingBackend, TabuBackend

backend = SimulatedAnnealingBackend()
```

Now we can use Quantum launcher to launch D-wave solver on the selected backend 

```python
from quantum_launcher.launcher import QuantumLauncher

launcher = QuantumLauncher(pr,alg,backend)
result = launcher.run()
result
```

The results corresponds to the following matching, since the solver is non-deterministic the result might be suboptimal:

```python
pr.visualize(result.best_bitstring)
```
