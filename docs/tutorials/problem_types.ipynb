# Problem Types

Quantum Launcher provides definitions and conversions to different models for a set of common combinatorial problems.

```python
from quantum_launcher.problems import MaxCut,EC,QATM,JSSP,TSP, GraphColoring
import os
```

Problems are equipped with various utility functions like loading from preset:

```python
ec = EC.from_preset("toy",onehot="exact")
```

drawing a visualization of the problem:

```python
ec.visualize()
```

drawing a visualization of a solution:

```python
ec.visualize("010101")
```

and generating a random problem instance:

```python
ec = EC.generate_ec_instance(5,10,0.4,onehot="exact")
ec.visualize()
```

All problems problems already implemented in the quantum launcher are presented below:
